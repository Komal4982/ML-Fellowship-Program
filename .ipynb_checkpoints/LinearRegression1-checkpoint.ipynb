```python
# outlier function

def remove_outliers(df):
    low=.05
    high=.95
    quantile_df=df.quantile([low,high])
    for name in list (df.columns):
        if ptypes.is_numeric_dtype(df[name]):
            df=df[(df[name]>quantile_df.loc[low,name])&(df[name]<quantile_df.loc[high,name])]
    return df
```

```python
df=remove_outliers(df)
```

```python
def feature_scaling(feature_data):
    for data in feature_data.columns:
            feature_data[data]=(feature_data[data]-feature_data[data].mean())/np.std(feature_data[data])
            return feature_data
df=feature_scaling(df)
# print(df)
```

```python
def Split(data):
    train_set=0.70*len(data)
    train=int(train_set)
    print("train data=",train)
    test_set=0.30*len(data)
    test=int(test_set)
    print("test data:",test)
        
    return train,test
```

```python
train,test=Split(df)
```
